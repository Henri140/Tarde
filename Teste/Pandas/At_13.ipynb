```python
#1
import pandas as pd
df = pd.read_excel('vinhos_exercicio.xlsx')

freqAbs = df['grape_type'].value_counts()
freqRel = df['grape_type'].value_counts(normalize=True) * 100
freqCum = freqAbs.cumsum()

tabGra = pd.DataFrame({
    'Frequência Absoluta': freqAbs,
    'Frequência Relativa (%)': freqRel,
    'Frequência Acumulada': freqCum
})

print(tabGra)
```

```python
#2
df['quartil'] = pd.qcut(df['value_usd'], q=4)

quartAbs = df['quartil'].value_counts().sort_index()
quartRel = df['quartil'].value_counts(normalize=True).sort_index() * 100

tabQuartis = pd.DataFrame({
    'Frequência Absoluta': quartAbs,
    'Frequência Relativa (%)': quartRel
})

print(tabQuartis)
```

```python
#3
bins = [2014, 2018, 2022, 2027]
labels = ['2014-2018', '2019-2022', '2023-2027']

df['year_group'] = pd.cut(df['year'], bins=bins, labels=labels, include_lowest=True)

yAbs = df['year_group'].value_counts().sort_index()
yRel = df['year_group'].value_counts(normalize=True).sort_index() * 100
yCum = yAbs.cumsum()

tabY = pd.DataFrame({
    'Frequência Absoluta': yAbs,
    'Frequência Relativa (%)': yRel,
    'Frequência Acumulada': yCum
})

print(tabY)
```
