```python
#1
import pandas as pd

df=pd.read_excel('vinhos_exercicio.xlsx')

md=df['volume_hl'].mean()
m=df.sample(frac=0.05,random_state=5)
me=m['volume_hl'].mean()
print(round(md,2),round(me,2))
```

```python
#2
if md> me:
    print(f'A amostra tem media maior que a população: {(me):.2f}')
else:
    print(f'A população tem media maior que a amostra: {(md):.2f}')
```

#3
Amostras podem variar porque estão sujeitas à variabilidade aleatória inerente ao processo de seleção, fazendo com que cada conjunto represente de forma ligeiramente diferente a população.
