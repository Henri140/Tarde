```python
#1
import pandas as pd
import numpy as np


df = pd.read_excel("vinhos_exercicio.xlsx")

df3 = df.copy()

for _ in range(20):
    i = np.random.randint(0, df3.shape[0])
    j = np.random.randint(0, df3.shape[1])
    df3.iat[i, j] = np.nan


df3 = pd.concat([df3, df3])


print("DF3 SHAPE:", df3.shape)
print(df3.describe())
```

```python
#2
null=df3.isnull().sum()
nullT=df3.isnull().sum().sum()

print(f'Númmeros nulos por colunas: {null}')
print('')
print(f'Total de números nulos: {nullT}')


'''Sem duplicadas'''
df4 = df3.drop_duplicates()

print("DF4 Shape:", df4.shape)
print(df4.describe())

df5 = df3.dropna()
print("Df5 Shape:", df5.shape)
print(df5.describe())

'''--Sem valores nulos'''
df5 = df3.dropna()

print("DF5 SHAPE:", df5.shape)
print(df5.describe())
```

```python
#3
print("antes:", df3.shape)

dfFinal = df3.drop_duplicates()

print("depois:", dfFinal.shape)

dfFinal["volume_hl"] = dfFinal["volume_hl"].fillna(dfFinal["volume_hl"].median())
dfFinal["value_usd"] = dfFinal["value_usd"].fillna(dfFinal["value_usd"].median())
```
