  Etapas do Projeto  
  1. Tratamento de Dados (Python)  
Utilizando Python no VS Code:

Carregar a base de dados

Identificar e tratar valores nulos

Remover duplicidades

Criar novas colunas (ex: preço médio)

Garantir a qualidade dos dados

```python
import pandas as pd
import numpy as np

df=pd.read_excel('vinhos_desafio.xlsx')

#limpeza de valores nulos
df['volume_hl'] = df['volume_hl'].apply(
    lambda x: np.random.choice(df['volume_hl'].dropna()) if pd.isnull(x) else x
)

df['valor_usd'] = df['valor_usd'].apply(
    lambda x: np.random.choice(df['valor_usd'].dropna()) if pd.isnull(x) else x
)

#remoção de duplicatas
df.drop_duplicates()

#criação da coluna preço medio
df['preco_medio']=df['valor_usd']/df['volume_hl']

#contagem de valores vazios por coluna
null=df.isnull().sum()
```

2. Análise ExploratóriaAinda em Python:
Identificar padrões nos dados
Comparar vendas, importação e exportação
Analisar evolução ao longo dos anos
Avaliar diferenças entre tipos de vinho

```python
#Correlação

print('--Correlação--')
corr = df.corr(numeric_only=True)
print(corr)

print('')

#Sazonalidade
print('--Sazonalidade--')
sazonalidade = df.groupby('ano')['volume_hl'].mean()
print(sazonalidade)
print('')

# Agrupamentos por tipo
print('--Agrupamento Tipo--')
uva = df.groupby('tipo_uva')['volume_hl'].sum()
categoria = df.groupby('categoria')['volume_hl'].sum()
local = df.groupby('local')['volume_hl'].sum()

print(uva)
print(categoria)
print(local)

print('')

# Comparação de categorias ao longo do tempo

print('--Comparação Categorias por Tempo--')
pivotUva = df.pivot_table(values='volume_hl', index='ano', columns='tipo_uva')
print(pivotUva)

print('')

# Evolução total

print('--Evolução Total--')
evolucaoT=df.groupby('ano')['volume_hl'].sum()
print(evolucaoT)

print('')

# Evolução por tipo de uva

print('--Evolução Tipo Uva--')
evolucaoUva = df.groupby(['ano','tipo_uva'])['volume_hl'].sum().unstack()
print(evolucaoUva)

print('')

# Crescimento percentual]

print('--Crescimento Percentual--')
crescimento = df.groupby('ano')['volume_hl'].sum().pct_change()
print(crescimento)

print('')

# Evolução detalhada

print('--Evolução em Detalhes--')
evolucao = df.groupby(['ano', 'tipo_uva'])['volume_hl'].sum().unstack()
print(evolucao)

print('')

# Comparação de médias

print('--Comparação de Médias--')
mediaCategoria = df.groupby('categoria')[['volume_hl','valor_usd','preco_medio']].mean()
print(mediaCategoria)

print('')

# Participação de mercado

print('--Participação no Mercado--')
participacao = df.groupby('tipo_uva')['volume_hl'].sum() / df['volume_hl'].sum()
print(participacao)

print('')

df.to_excel('VinhoTratado.xlsx')
```
