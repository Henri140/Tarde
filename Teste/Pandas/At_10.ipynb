```python
#1
import pandas as pd

df=pd.read_excel("vinhos_exercicio.xlsx")

dfW=df[
    (df["grape_type"]== "Wine") &
    (df["location_type"] == "Brazil_State") &
    (df["category"] == "Sales")
    ]

dfT=df[
    (df["grape_type"] == "Table") &
    (df["category"]== "Sales")
]

descW = dfW.describe().T
descT = dfT.describe().T

print(descW)
print(descT)

mediaW = descW.loc["value_usd", "mean"]
mediaT = descT.loc["value_usd", "mean"]

if mediaW > mediaT:
    print("Wine possui maior média de value_usd")
else:
    print("Table possui maior média de value_usd")
```

```python
#2
dfWs = dfW.sort_values(by=["year", "value_usd"], ascending=[False, True])
dfTs = dfT.sort_values(by=["year", "value_usd"], ascending=[False, True])


dfWf = dfWs.iloc[:, [df.columns.get_loc(col) for col in ["year", "grape_type", "volume_hl", "value_usd", "location"]]]
dfTf= dfTs.iloc[:, [df.columns.get_loc(col) for col in ["year", "grape_type", "volume_hl", "value_usd", "location"]]]

print(dfWf.head())
print(dfTf.head())
```

```python
#3
dfWf.reset_index(drop=True, inplace=True)
dfTf.reset_index(drop=True, inplace=True)

dfW10 = dfWf.loc[:9]
dfT10 = dfTf.loc[:9]

print(dfW10)
print(' ')
print(dfT10)
```
