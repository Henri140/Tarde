```python
def contar_pares(*n):
    i=0
    for n in n:
        if n%2==0:
            i+=1
    print(f'Tem {i} números pares entre os números digitados')

contar_pares(2,10)
```

```python
def cubica(*nums):
    for nums in nums:
        nl=nums
        rs=nl**3
        print(f'{nums}³ = {rs}')

cubica(1,2,3,4,5)
```

```python
def  lista_compras(**produtosValor):
    rs=0

    for produto,valor in produtosValor.items():
        print(f'Produto:{produto} Valor:{valor}')
        rs+=valor
    
    print(f'Total: {rs}')

lista_compras(Macarrão=20.50,Frango=52.50)
```

```python
def divisao(a, b):
    try:
        rs = a / b
    except ZeroDivisionError:
        print('Digite um número diferente de zero')
    else:
        print(f'Conta realizada com sucesso: {rs}')
    finally:
        print('Fim da operação')

divisao(10, 0)
```

```python
def caixa(nomeP):
    mercado = {
        'Macarrão': 16.20,
        'Molho': 2.75,
        'Azeitona': 23.99
    }

    try:
        preco = mercado[nomeP]
        print(f'O {nomeP} custa: {preco}')
    except KeyError:
        print('Produto não encontrado')
    finally:
        print(f'Consulta encerrada')

caixa('Macarrão')
```
